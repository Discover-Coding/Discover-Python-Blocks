# 2.2. Python Random

This lets us use some random *functions* to generate a random number. There are [a lot to choose from](https://docs.python.org/3/library/random.html), but here are 2 very common functions:

- `random.random()` picks a *DECIMAL* number between `0` and `1`
- `random.randint(x,y)` picks a *WHOLE* number between `x` and `y`

# 5. Sample Code for Base Game


````python
import random

# Get some input from the player using the input() function
choice = input("Choose 1 for rock, 2 for paper or 3 for scissors:")
player = int(choice)

# print what the Player chooses
if player == 1 :
    print ("Player chooses rock")
elif player == 2 :
    print ("Player chooses paper")
else :
    print ("Player chooses scissors")
    
# CPU chooses a random number between 1 and 3 using random.randint(x,y)
cpu = random.randint(1, 3)

# print what the CPU chooses
if cpu == 1 :
    print ("CPU chooses rock")
elif cpu == 2 :
    print ("CPU chooses paper")
else :
    print ("CPU chooses scissors")

# If the player plays ROCK, check what the CPU plays
if player == 1 :
    if cpu == 1 : 
        print ("TIE")
    elif cpu == 2 :
        print ("CPU wins")
    else :
        print ("Player wins")
        
# If the player plays PAPER, check what the CPU plays and print TIE, CPU wins or Player wins
if player == 2 :
    if cpu == 1 : 
        print ("Player wins")
    elif cpu == 2 :
        print ("TIE")
    else :
        print ("CPU wins")
        
# If the player plays SCISSORS, check what the CPU plays
if player == 3 :
    if cpu == 1 : 
        print ("CPU wins")
    elif cpu == 2 :
        print ("Player wins")
    else :
        print ("TIE")
````

# 6. Sample code for Advanced Game Algorithm

````python
import random

game = 1
player_score = 0
cpu_score = 0

while game <= 3 : 
    # Get some input from the player using the input() function
    choice = input("Choose 1 for rock, 2 for paper or 3 for scissors:")
    player = int(choice)

    # print what the Player chooses
    print("Player chooses:")
    if player == 1 :
        print(r"""
               .----.
              / ~   \\
              \  ~   /
               `----'
                """)
    elif player == 2 :
        print(r"""
                ______
               /     /
              /     /
             /_____/
             """)
    else :
        print(r"""
              \ /
               X
              O O
             """)
    # CPU chooses a random number between 1 and 3 using random.randint(x,y)
    cpu = random.randint(1, 3)

    # print what the CPU chooses
    print("CPU chooses:")
    if cpu == 1 :
        print(r"""
               .----.
              / ~   \\
              \  ~   /
               `----'
                """)
    elif cpu == 2 :
        print(r"""
                ______
               /     /
              /     /
             /_____/
             """)
    else :
        print(r"""
              \ /
               X
              O O
             """)

    # If the player plays ROCK, check what the CPU plays
    if player == 1 :
        if cpu == 1 : 
            print ("TIE")
        elif cpu == 2 :
            print ("CPU wins")
            cpu_score = cpu_score + 1
        else :
            print ("Player wins")
            player_score = player_score + 1

    # If the player plays PAPER, check what the CPU plays and print TIE, CPU wins or Player wins
    if player == 2 :
        if cpu == 1 : 
            print ("Player wins")
            player_score = player_score + 1
        elif cpu == 2 :
            print ("TIE")
        else :
            print ("CPU wins")
            cpu_score = cpu_score + 1

    # If the player plays SCISSORS, check what the CPU plays
    if player == 3 :
        if cpu == 1 : 
            print ("CPU wins")
            cpu_score = cpu_score + 1
        elif cpu == 2 :
            print ("Player wins")
            player_score = player_score + 1
        else :
            print ("TIE")
            
    game = game + 1
    
print( "Player score = " + str(player_score) )
print( "CPU score = " + str(cpu_score) )
````
